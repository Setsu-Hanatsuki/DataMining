```python
import pandas as pd
df=pd.read_csv("sample_custmer.csv",encoding="shift-jis")
df
```

```python
dum=pd.get_dummies(df,columns=["SEX","AD"])
dum
```

```python
dum.columns
```

```python
y_name='CV'
y=dum[y_name].values
x_table=dum.drop([y_name,'ID'],axis=1)
name=x_table.columns
x=x_table.values
print(y)
```

```python
class_name=["no","yes"]
y2=[]
for i in range(len(y)):
    if y[i]=='yes':
        y2.append(1)
    else:
        y2.append(0)
print(y2)
print(x)
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
```

```python
model=DTC()
```

```python
model.fit(x,y2)
```

```python
import numpy as np
imp=model.feature_importances_
out=[]
out.append(name)
out.append(imp)
out=np.array(out).T
dfo=pd.DataFrame(out)
dfo.columns=["項目","重要度"]
dfo
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
dotdata=export_graphviz(model,filled=True,rounded=True,class_names=class_name,feature_names=name,out_file=None)
graph=graph_from_dot_data(dotdata)
graph.write_png("tree.png")
image=np.array(Image.open("tree.png"))
plt.imshow(image)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import preprocessing
x2=preprocessing.minmax_scale(x)
score="f1"
tuned_parameters = [{'C': [0.1 ,1, 10, 100, 1000]}]
SVM=GridSearchCV(SVC(),tuned_parameters,cv=10,scoring='%s_weighted' % score)
SVM.fit(x2,y2)
SVM.best_params_
```

```python
model2=SVC(kernel='linear',C=10)
```

```python
model2.fit(x2,y2)
```

```python
model2.coef_
```

```python
import matplotlib.pyplot as plt
import numpy as np
k=np.arange(0,len(name),1)
plt.bar(k,model2.coef_[0],tick_label=name)
plt.show()
```
