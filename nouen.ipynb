```python
dfd3=dfd2.drop("季節",axis=1)
dfd3
```

```python
y2=dfd3["燃焼回数回数"].values
dfd3_x=dfd3.drop("燃焼回数回数",axis=1)
name_rx=dfd3_x.columns
x3=dfd3_x.values
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x3,y2)
```

```python
a=model.coef_
b=model.intercept_
y_pred=model.predict(x3)
```

```python
se=np.sum((y-y_pred)**2,axis=0)
se=se/(x3.shape[0]-x3.shape[1]-1)
s=np.linalg.inv(np.dot(x3.T,x3))
stder=np.sqrt(np.diagonal(se*s))
```

```python
t=a/stder
```

```python
out4=[]
out4.append(name_rx)
out4.append(a)
out4.append(t)
out4=np.array(out4).T
dfo3=pd.DataFrame(out4)
dfo3.columns=[" 因子名" ,"重み ","t値"]
dfo3
```

```python
dfo3.to_csv("重回帰燃焼回数.csv",encoding="shift-jis")
```
