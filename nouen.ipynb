```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x3,y2)
```

```python
a=model.coef_
b=model.intercept_
y_pred=model.predict(x3)
```

```python
se=np.sum((y-y_pred)**2,axis=0)
se=se/(x3.shape[0]-x3.shape[1]-1)
s=np.linalg.inv(np.dot(x3.T,x3))
stder=np.sqrt(np.diagonal(se*s))
```

```python
t=a/stder
```

```python
out4=[]
out4.append(name_rx)
out4.append(a)
out4.append(t)
out4=np.array(out4).T
dfo3=pd.DataFrame(out4)
dfo3.columns=[" 因子名" ,"重み ","t値"]
dfo3
```

```python
print("切片"+str(b))
print("決定係数"+str(model.score(x3,y2)))
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=len(x2[0]))
pca.fit(x2)
#寄与率
con=pca.explained_variance_ratio_

#固有ベクトル
fac=pca.components_
```

```python
table=[]
tmp=[]
for i in range(len(con)):
    tmp.append("第"+str(i+1)+"主成分")
    for j in range(len(fac[i])):
        tmp.append(fac[i][j])
    tmp.append(con[i])
    table.append(tmp)
    tmp=[]
names=[]
names.append("主成分軸")
for i in range(len(name)):
    names.append(name[i])
names.append("寄与率")
dfpca=pd.DataFrame(table)
dfpca.columns=names
dfpca
```

```python
dfpca.to_csv("主成分.csv",encoding="shift-jis",index=False)
```
